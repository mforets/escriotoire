## In for solutions

```julia
using Revise, ReachabilityAnalysis, Plots
```

```julia
A = [0 1; -1 0.]
prob = @ivp(x' = Ax, x(0) ∈ BallInf(ones(2), 0.1))
sol = solve(prob, tspan=(0.0, 7.0));
```

```julia
plot(sol, vars=(1, 2), ratio=1.)
```

```julia
@which [0.0, 0.0] ∈ set(sol[1]) # LazySets
```

```julia
@which [0.0, 0.0] ∈ sol[1] # ReachSet
```

```julia
@which [0.0, 0.0] ∈ flowpipe(sol) # of the whole flowpipe
```
