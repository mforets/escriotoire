```julia
using ONSAS, Plots
```

```julia
# problem formulation
M = [2 0; 0 1.]
K = [6 -2; -2 4.]
C = zeros(2, 2)
R = [0, 10.]
sys = SecondOrderAffineContinuousSystem(M, C, K, R)

U₀ = zeros(2)
U₀′ = zeros(2)
NSTEPS = 10
prob = InitialValueProblem(sys, (U₀, U₀′));
```

```julia
# solution
NSTEPS = 2000
sol = solve(prob, CentralDifference(δ=0.01), NSTEPS);
```

```julia
plot([U[1] for U in sol], lab="")
```
