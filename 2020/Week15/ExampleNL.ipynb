## Flowpipe computation using a Taylor model approach

```julia
using ReachabilityAnalysis, Plots
```

```julia
@taylorize function f(du, u, p, t)
    du[1] = u[3] * u[1] - u[4] * (u[1] * u[2]) - u[7] * u[1]^2
    du[2] = -u[5] * u[2] + u[6] * (u[1] * u[2])
    
    # encode uncertain params
    du[3] = zero(u[1]) # p[1]
    du[4] = zero(u[1]) # p[2]
    du[5] = zero(u[1]) # p[3]
    du[6] = zero(u[1]) # p[4]
    du[7] = zero(u[1]) # p[5]
end
```

```julia
# encode initia-value problem
p_int = (0.99..1.01) × (0.99..1.01) × (2.99..3.01) × (0.99..1.01) × (0.099..0.101)
U0 = Singleton([1.0, 1.0]) × convert(Hyperrectangle, p_int)
prob = @ivp(u' = f(u), dim: 7, u(0) ∈ U0);
```

```julia
@time sol = solve(prob, tspan=(0.0, 10.0));
```

```julia
sol.alg.abs_tol
```

```julia
sol_zono = overapproximate(sol, Zonotope);
typeof(sol_zono)
```

```julia
plot(sol_zono, vars=(1, 2), lw=0.3, title="Uncertain params", lab="abs_tol = 1e-15", xlab="u1", ylab="u2")
```

## Changing the tolerance

The approximation is more coarse, but it's faster if we reduce the tolerance in the `TMJets` algorithm:

```julia
@time sol = solve(prob, tspan=(0.0, 10.0), TMJets(abs_tol=1e-10));
sol_zono = overapproximate(sol, Zonotope);
plot!(sol_zono, vars=(1, 2), lw=0.3, color=:red, alpha=0.2, lab="abs_tol = 1e-10")
```

## Uncertain initial condition (u0)

```julia
# now we consider an initial box around u0
u0 = Singleton([1.0, 1.0])
□(ϵ) = BallInf(zeros(2), ϵ)
U0 = (u0 ⊕ □(0.05)) × convert(Hyperrectangle, p_int)

prob = @ivp(u' = f(u), dim: 7, u(0) ∈ U0)

@time sol = solve(prob, tspan=(0.0, 10.0), TMJets(abs_tol=1e-10))
sol_zono = overapproximate(sol, Zonotope)
plot(sol_zono, vars=(1, 2), color=:orange, lw=0.3,
     lab="eps = 0.05", title="Uncertain u0 and uncertain params",
     xlab="u1", ylab="u2")
```

```julia
# now we consider an initial box around u0
U0 = (u0 ⊕ □(0.01)) × convert(Hyperrectangle, p_int)
prob = @ivp(u' = f(u), dim: 7, u(0) ∈ U0)

@time sol = solve(prob, tspan=(0.0, 10.0), TMJets(abs_tol=1e-10))
sol_zono = overapproximate(sol, Zonotope)
plot!(sol_zono, vars=(1, 2), color=:blue, lw=0.3,
     lab="eps = 0.01", title="Uncertain u0 and uncertain params",
     xlab="u1", ylab="u2")
```
