```julia
using LazySets, Plots
```

## Random sampling from a hyperrectangle

```julia
X = rand(Hyperrectangle)
```

```julia
S = sample(X, 1000);
```

```julia
plot(X)
```

```julia
plot!(Singleton.(S))
```

## Random sampling from a flat hyperrectangle

```julia
X = convert(Hyperrectangle, Interval(0, 0) × Interval(0, 1))
```

```julia
plot(X)
```

```julia
S = sample(X, 100);
```

```julia
plot!(Singleton.(S))
```

## Uniform split

```julia
X = rand(Hyperrectangle)
S = split(X, [10, 10])

plot(X)
plot!(S)
```

```julia
# get all vertices
V = reduce(vcat, vertices_list.(S));

plot(X)
plot!(Singleton.(V))
```

## Non-uniform split

```julia
X = rand(Hyperrectangle)
S = split(X, [5, 20])

plot(X)
plot!(S)
```
