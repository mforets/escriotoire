## Exact projection of a Taylor model reach-set

$$
T =  \{y : y = p_0(x) + p_1(x)t + \ldots + p_k(x)t^k + w: x \in [-1, 1]^n, w \in R, t \in \Delta t\}
$$

Projection onto $y_1$ means to set $y_2 = y_3 = \ldots = 0$:

- $y_1 = p_0^{(1)}(x) + p_1^{(1)}(x)t + \ldots + p_k^{(1)}(x)t^k$

- $0 = p_0^{(2)}(x) + p_1^{(2)}(x)t + \ldots + p_k^{(2)}(x)t^k$

- $\vdots$

- $0 = p_0^{(n)}(x) + p_1^{(n)}(x)t + \ldots + p_k^{(n)}(x)t^k$

for all $x \in [-1, 1]^n$
