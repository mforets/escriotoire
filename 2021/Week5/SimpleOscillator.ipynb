```julia
using ReachabilityAnalysis, Plots
```

```julia
# initial-value problem for the harmonic oscillator with a box initial condition
A = [0 1; -1 0.]
X0 = BallInf(ones(2), 0.2)
prob = @ivp(x' = Ax, x(0) ∈ X0)

# compute the flowpipe
sol = solve(prob, tspan=(0.0, 8.0), alg=GLGM06(δ=0.05));

# "bad" states
B = HalfSpace([0.0, -1.0], -1.5);
```

```julia
fig = plot(xlab="x(t)", ylab="y(t)", ylims=(-2, 2), ratio=1., legend=:bottomleft)

plot!(fig, sol, vars=(1, 2))
plot!(fig, B, lab="Bad states", c=:red)
plot!(fig, X0, lab="Initial states", c=:orange)
```

```julia
fig = plot(xlab="t", ylab="y(t)", xlims=(0.0, 8.0))

plot!(fig, sol, vars=(0, 2))
plot!(B, c=:red, lab="Bad states")
```
